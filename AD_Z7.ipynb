Napisz algorytm k-mean w wersji Lloyd (klasyczny). Nie można używać żadnych funkcji z k-means w pythonie. 

  * wszysko piszemy ręcznie
  * uzupełnij implementację poniższej klasy

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import random

class KMeans(object):
    """
    """
    def __init__(self):
        
        self.labels = None
        self.clusters = None
        self.n_clusters = None
        self.cluster_centers = None
        

    def fit(self, X, n_clusters):
        self.X = X
        self.labels = [i for i in range(n_clusters)]
#         self.cluster_centers = [np.random.sample(size=X[0].shape) for _ in range(n_clusters)]
#         old_mean = [np.random.sample(size=X[0].shape) for _ in range(n_clusters)]
        self.cluster_centers = [np.random.sample(size=X[0].shape)*8 for _ in range(n_clusters)]
        old_mean = [np.random.sample(size=X[0].shape)*8 for _ in range(n_clusters)]

        while not self.stop(self.cluster_centers, old_mean):
            old_mean = self.cluster_centers
            self.cluster_points(X, self.cluster_centers)
            self.change_centers()

    
    def predict(self, X):
        result = []
        for x in X:
            result.append(self.predict_point(x))
        return result
    
    def predict_point(self, x):
        return min([(i[0], np.linalg.norm(x-self.cluster_centers[i[0]])) for i in enumerate(self.cluster_centers)], key=lambda t:t[1])[0]
    
    def cluster_points(self, X, mean):
        self.clusters  = {}
        for x in self.X:
            best_key = min([(i[0], np.linalg.norm(x-mean[i[0]])) for i in enumerate(mean)], key=lambda t:t[1])[0]
            try:
                self.clusters[best_key].append(x)
            except KeyError:
                self.clusters[best_key] = [x]
        return self.clusters
 
    def change_centers(self):
        keys = sorted(self.clusters.keys())
        self.cluster_centers = []
        for k in keys:
            self.cluster_centers.append(np.mean(self.clusters[k], axis = 0))
    
    def stop(self, mean, old_mean):
#         print (set([tuple(a) for a in mean]), set([tuple(a) for a in old_mean]))
        return (set([tuple(a) for a in mean]) == set([tuple(a) for a in old_mean]))

    def debug(self):
        print(list( (type(key), key, len(value)) for key,value in self.clusters.items()))
```

# Zadanie

Sprawdż działanie napisanego algorytmu na poniższych danych.

```python
from sklearn.datasets.samples_generator import make_blobs
X_blobs, y_blobs = make_blobs(n_samples=1000, centers=3, n_features=2, random_state=0)
n_classes = len(set(y_blobs))
plt.scatter(X_blobs[:, 0], X_blobs[:, 1], c=y_blobs)
plt.show()
```

```python
def plot_kmeans(k_means, colors=['red','blue','green']):
    for k,v in k_means.clusters.items():
        data = np.array(v)
        plt.scatter(data[:, 0], data[:, 1], c=colors[k])
    plt.show()
    
k_means = KMeans()
k_means.fit(X_blobs, 3)
plot_kmeans(k_means)
```
