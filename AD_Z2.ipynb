## Przykład
Użyj testu kstest do obliczenia wartości p-value.

```python
x = rnd.randn(1000)
plt.hist(x)
plt.show()
kstest = stats.kstest
stat, pval = kstest(x, 'norm')
print(stat)
print(pval)
if(pval < 0.05):
    print("Odrzucam hipotezę zerową na poziomie istotności 0.05")
else:    
    print("Nie ma podstaw by odrzucić hipotezę zerową  na poziomie istotności 0.05")
```
