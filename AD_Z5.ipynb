# Zadanie 1
Wygeneruj dane 2d oraz 3d w kształcie prostokąta (odpowiednio sześcianu). Wykonaj PCA i narysuj wszystkie (odpowiednio dwa i trzy) komponenty.

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

data = np.column_stack((np.random.uniform(0, 1, 1000), np.random.uniform(0, 1, 1000)))
def plot_points(data):
    plt.scatter(data[:,0], data[:,1])
    plt.legend(loc=2)
    
plot_points(data)
pca = PCA(n_components=2)
pca.fit(data)
res = pca.transform(data)
print(pca.components_)


# TUTAJ PLOT PO PROTU WEKTORUY KTORE WYSZLY!!!!
transformed_data = pca.transform(data)
first_pc = pca.components_[0]
second_pc = pca.components_[1]
for ii, jj in zip(transformed_data, data):
    plt.scatter(first_pc[0]*ii[0], first_pc[1]*ii[0], color='r')
    plt.scatter(second_pc[0]*ii[1], second_pc[1]*ii[1], color='c')
    plt.scatter(jj[0], jj[1], color='b')
plt.show()

````

````python
data3 = np.column_stack((np.random.uniform(0, 1, 1000), np.random.uniform(0, 1, 1000), np.random.uniform(0, 1, 1000)))
pca3 = PCA(n_components=2)
pca3.fit(data)
transformed_data3 = pca3.transform(data)
first_pc = pca3.components_[0]
second_pc = pca3.components_[1]
for ii, jj in zip(transformed_data3, data3):
    plt.scatter(first_pc[0]*ii[0], first_pc[1]*ii[0], color='r')
    plt.scatter(second_pc[0]*ii[1], second_pc[1]*ii[1], color='c')
    plt.scatter(jj[0], jj[1], color='b')
plt.show()
````
