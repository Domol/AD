## Zadanie 1
Wykonaj algorytm k-menas na poni≈ºszych danych z $k=2$.

````python
import matplotlib.pyplot as plt
import numpy as np
import numpy.random as rnd
from sklearn.cluster import KMeans
import numpy as np

rnd.seed(1)
n = 30
points = np.concatenate([rnd.multivariate_normal((4,15),np.array([[1,0],[0,30]]),n),
                    rnd.multivariate_normal((10,20),np.array([[1,0],[0,30]]),n)],axis=0)
labels = np.repeat([0,1],n)


kmeans = KMeans(n_clusters=2, init='random')
k = kmeans.fit(points)
print(k.cluster_centers_ )



plt.scatter(points[:,0],points[:,1],c = np.array(["red","green", "blue"])[labels])
plt.scatter(k.cluster_centers_[:,0], k.cluster_centers_[:,1], c="b")
plt.xlim(0,30)
plt.ylim(0,30)
plt.show()
````

# Zadanie 3
Narysuj dane iris data wybierajƒÖc dwie pierwsze wsp√≥≈Çrzƒôdne.
Wykonaj klastrowanie algorytmem k-means na danych iris:

  * bez skalowania i $k=3$
  * ze skalowaniem i $k=3$ 

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
from sklearn import preprocessing
%matplotlib inline

# import some data to play with
iris = datasets.load_iris()


X = iris.data[:, :2]  # we only take the first two features.
y = iris.target

def test_kmeans(X, y, n):
    kmeans_iris = KMeans(n_clusters=n, init='random')
    ki = kmeans_iris.fit(X)


    plt.scatter(X[:, 0], X[:, 1], c=y)
    plt.scatter(ki.cluster_centers_[:,0], ki.cluster_centers_[:,1], c="red")
    plt.show()
test_kmeans(X, y, 3)

scaler = StandardScaler()
scaler.fit(X)
new_X = scaler.transform(X)


test_kmeans(new_X, y, 3)
````

# Zadanie 4
Wykonaj klastrowanie k-means na danych iris z ilo≈õciƒÖ grup (k=1,2,3,4,5,6) oraz narysuj wykres jak zmienia siƒô suma kwadrat√≥w odleg≈Ço≈õci od centr√≥w grup (Sum of distances of samples to their closest cluster center) w zale≈ºno≈õci od ilo≈õci grup.

````python
iris = datasets.load_iris()


X = iris.data[:, :2]  # we only take the first two features.
y = iris.target

ks = range(1,7)
sums_of_distances = []
for k in ks:
    kmeans = KMeans(n_clusters=k, init='random')
    k = kmeans.fit(X)
    sums_of_distances.append(k.inertia_)
    
plt.plot(ks, sums_of_distances)
plt.show()
````

# Zadanie 5
Wykonaj klastrowanie k-means na danych iris z ilo≈õciƒÖ grup (k=1,2,3,4,5,6) oraz oce≈Ñ ich jako≈õƒá za pomocƒÖ miary Adjusted Rand index
http://scikit-learn.org/stable/modules/clustering.html#clustering-evaluation.

````python
from sklearn import metrics
iris = datasets.load_iris()


X = iris.data[:, :2]  # we only take the first two features.
y = iris.target
def test_kmeans_adjuster_rand_index(X, y, ks):
    results = []
    for k in ks:
        kmeans = KMeans(n_clusters=k, init='random')
        k = kmeans.fit(X)
        labels_pred = kmeans.labels_
        results.append(metrics.adjusted_rand_score(y, labels_pred))
    plt.plot(ks, results)
    plt.show()
test_kmeans_adjuster_rand_index(X, y, range(1,7))
````

````python

def image_segmentation(filename, n_clusters, scaled=False):
    im = imread(filename, flatten=False, mode='RGB')
    data = []
    for i in range(im.shape[0]):
        for j in range(im.shape[1]):
            data.append(np.concatenate((im[i,j],np.array([i,j]))))
    data = np.array(data)
    
    if scaled:
        scaler = StandardScaler()
        scaler.fit(data)
        data = scaler.transform(data)
        print(data)
        
    kmeans = KMeans(n_clusters=n_clusters, init='random')
    k = kmeans.fit(data)
    plt.imshow(k.labels_.reshape(im.shape[:2]))
    plt.show()
    
image_segmentation('pigeon.jpg', 6)
image_segmentation('pigeon.jpg', 6, scaled=True)
````

````python
import matplotlib.pyplot as plt
def print_faces(images, target, top_n):
    # set up the figure size in inches
    fig = plt.figure(figsize=(12, 12))
    fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)
    for i in range(top_n):
        # plot the images in a matrix of 20x20
        p = fig.add_subplot(20, 20, i + 1, xticks=[], yticks=[])
        p.imshow(images[i], cmap=plt.cm.bone)

        # label the image with the target value
        p.text(0, 14, str(target[i]))
        p.text(0, 60, str(i))
    plt.show()
````
